```python
## create table Venue
%%sql 
DROP TABLE IF EXISTS Venue CASCADE;
CREATE TABLE Venue (
    vid int,
    PRIMARY KEY (vid),
    venue_name text,
    venue_address text,
    state text,
    zip_code text)
```

```python
## create table Artist
%%sql
DROP TABLE IF EXISTS Artists CASCADE;
CREATE TABLE Artists (
    aid int,
    name text,
    description text, 
    video text, 
    PRIMARY KEY (aid)
)
```

```python
## create table Event Organizers
%%sql
DROP TABLE IF EXISTS Event_Organizers CASCADE;
CREATE TABLE Event_Organizers (
    eoid int,
    name text,
    PRIMARY KEY (eoid)
)
```

```python
## create table Events 
%%sql
DROP TABLE IF EXISTS Events CASCADE;
CREATE TABLE Events (
  eid int, 
  eoid int, 
  events_name text,
  PRIMARY KEY (eid, eoid), 
  FOREIGN KEY (eoid) REFERENCES Event_Organizers 
  ON DELETE CASCADE
)
```

```python
## create table Events_Venue for the relation Happen
%%sql 
DROP TABLE IF EXISTS Events_Venue CASCADE;
CREATE TABLE Events_Venue (
  eid int, 
  eoid int, 
  vid int NOT NULL, 
  PRIMARY KEY (eid, eoid), 
  FOREIGN KEY (vid) REFERENCES Venue,
  FOREIGN KEY (eid, eoid) REFERENCES Events
)
```

```python
## create table Artists_Events for the relation Perform
%%sql 
DROP TABLE IF EXISTS Events_Artists CASCADE;
CREATE TABLE Events_Artists (
  eid int,
  aid int NOT NULL,
  event_date date,
  PRIMARY KEY (eid, aid),
  FOREIGN KEY (aid) REFERENCES Artists,
  FOREIGN KEY (eid) REFERENCES Artists  
  ON DELETE NO ACTION
)
```

```python
## create table Tickets
%%sql
DROP TABLE IF EXISTS Tickets CASCADE;
CREATE TABLE Tickets (
  tid int, 
  eid int, 
  eoid int,
  price float(2),
  seat_number text,
  is_resold boolean,
  resold_price float(2),
  PRIMARY KEY (tid), 
  FOREIGN KEY (eid, eoid) REFERENCES Events 
  ON DELETE CASCADE
)
```

```python
## create table Users
%%sql 
DROP TABLE IF EXISTS Users CASCADE;
CREATE TABLE Users (
    uid int,
    user_name text,
    user_address text,
    PRIMARY KEY (uid)
)
```

```python
## create table User_Buy_Tickets
%%sql 
DROP TABLE IF EXISTS User_Buy_Tickets CASCADE;
CREATE TABLE User_Buy_Tickets (
  uid int,
  tid int, 
  eid int,
  eoid int,
  order_details text,
  buy_date date,
  PRIMARY KEY (eoid, eid, tid),
  FOREIGN KEY (uid) REFERENCES Users,
  FOREIGN KEY (eoid, eid, tid) REFERENCES Tickets
)
```

```python
## create table User_Resell_Tickets
%%sql 
DROP TABLE IF EXISTS User_Resell_Tickets CASCADE;
CREATE TABLE User_Resell_Tickets (
  uid int,
  tid int,
  eid int,
  eoid int,
  previous_oid int,
  resell_oid int,
  resell_details text, 
  resell_date date,
  PRIMARY KEY (eoid, eid, tid),
  FOREIGN KEY (uid) REFERENCES Users,
  FOREIGN KEY (eoid, eid, tid) REFERENCES Tickets
)
```

```python
## create table User_Refund_Tickets
%%sql 
DROP TABLE IF EXISTS User_Refund_Tickets CASCADE;
CREATE TABLE User_Refund_Tickets (
  uid int,
  tid int,
  eid int,
  eoid int,
  refund_id int,
  refund_date date,
  refund_details text,
  PRIMARY KEY (eoid, eid, tid),
  FOREIGN KEY (uid) REFERENCES Users,
  FOREIGN KEY (eoid, eid, tid) REFERENCES Tickets
)
```
