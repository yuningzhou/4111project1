## Application-level features enabled by our expansion

We added the application of reviews and ratings. Users can write reviews for any event as well as rate the event. Also users can like the reviews wrote by others. \
Users can also share the page of an event to another platform such as twitter, instagram etc. \
Moreover, we added the application of merchandise, such as T-shirts, posters, etc. Users can buy the merchandises of any event. 

## Description of the extensions on the database design

New entity sets: Review, Merchandise \
New relation sets: Like, Write, Relate, Rate, Share, Purchase, Have \
Description: \
**User** can write **Review** to **Events**. One user can write multiple reviews, but one review can be write by exacly one user. One event can have multiple reviews, but one review can only belong to exactly one event. \
**User** can **Like** **Review**. One user can like multiple reviews, and one review can be liked by multiple users. \
**User** can **Rate** **Events**. One user can rate multiple events and one event can be rated by multiple users. \
**User** can **Share** **Events** to other platforms. One user can share multipe events to other platforms, and one event can be shared by multiple users. \
**User** can purchase **Merchandise** of **Events**. One user can purchase multiple merchandises and one kind of merchandise can be perchased by multiple users. An event can have multiple merchandises and one merchandise can belong to multiple events.

Create Schema

```python
## create table Venue
%%sql 
DROP TABLE IF EXISTS Venue CASCADE;
CREATE TABLE Venue (
    vid int,
    PRIMARY KEY (vid),
    name text,
    address text, 
    city text,
    state text
    )
```

```python
## create table Artist
%%sql
DROP TABLE IF EXISTS Artists CASCADE;
CREATE TABLE Artists (
    aid int,
    name text,
    description text,  
    PRIMARY KEY (aid)
)
```

```python
## create table Event Organizers
%%sql
DROP TABLE IF EXISTS Event_Organizers CASCADE;
CREATE TABLE Event_Organizers (
    eoid int,
    name text,
    PRIMARY KEY (eoid)
)
```

```python
## create table Events_Venue 
%%sql
DROP TABLE IF EXISTS Event_Venue CASCADE;
CREATE TABLE Event_Venue (
  eid int, 
  eoid int, 
  vid int NOT NULL, 
  events_name text,
  PRIMARY KEY (eid, eoid), 
  FOREIGN KEY (eoid) REFERENCES Event_Organizers,
  FOREIGN KEY (vid) REFERENCES Venue
  ON DELETE NO ACTION
)
```

```python
## create table Artists_Events for the relation Perform
%%sql 
DROP TABLE IF EXISTS Events_Artists CASCADE;
CREATE TABLE Events_Artists (
  eid int, 
  eoid int, 
  aid int NOT NULL,
  event_date date,
  PRIMARY KEY (eid, eoid, aid),
  FOREIGN KEY (aid) REFERENCES Artists,
  FOREIGN KEY (eid, eoid) REFERENCES Event_Venue
)
```

```python
## create table Tickets
%%sql
DROP TABLE IF EXISTS Tickets CASCADE;
CREATE TABLE Tickets (
  tid int, 
  eid int NOT NULL, 
  eoid int NOT NULL,
  price float(2),
  seat_number text,
  PRIMARY KEY (tid), 
  FOREIGN KEY (eid, eoid) REFERENCES Event_Venue 
)
```

```python
## create table Users
%%sql 
DROP TABLE IF EXISTS Users CASCADE;
CREATE TABLE Users (
    uid int,
    user_name text,
    email text,
    phone text,
    user_address text, 
    state text,
    zip_code text,
    PRIMARY KEY (uid)
)
```

```python
## create table User_Buy_Tickets
%%sql 
DROP TABLE IF EXISTS User_Buy_Tickets CASCADE;
CREATE TABLE User_Buy_Tickets (
  uid int,
  tid int,
  buy_date date,
  PRIMARY KEY (uid, tid),
  FOREIGN KEY (uid) REFERENCES Users,
  FOREIGN KEY (tid) REFERENCES Tickets
)
```

```python
## create table User_Resell_Tickets
%%sql 
DROP TABLE IF EXISTS User_Resell_Tickets CASCADE;
CREATE TABLE User_Resell_Tickets (
  uid int,
  tid int, 
  resell_price float(2),
  resell_date date,
  PRIMARY KEY (uid, tid),
  FOREIGN KEY (uid) REFERENCES Users,
  FOREIGN KEY (tid) REFERENCES Tickets
)
```

```python
## create table User_Refund_Tickets
%%sql 
DROP TABLE IF EXISTS User_Refund_Tickets CASCADE;
CREATE TABLE User_Refund_Tickets (
  uid int,
  tid int,
  refund_date date,
  PRIMARY KEY (uid, tid),
  FOREIGN KEY (uid) REFERENCES Users,
  FOREIGN KEY (tid) REFERENCES Tickets
)
```

```python
## create table Review
%%sql 
DROP TABLE IF EXISTS Reviews CASCADE;
CREATE TABLE Reviews (
  rid int,
  PRIMARY KEY(rid), 
  content text
)
```

```python
## create table User_Review
%%sql 
DROP TABLE IF EXISTS User_Review CASCADE;
CREATE TABLE User_Review (
  rid int, 
  uid int NOT NULL, 
  PRIMARY KEY(rid), 
  review_date date,
  FOREIGN KEY (uid) REFERENCES Users,
  FOREIGN KEY (rid) REFERENCES Reviews
)
```

```python
## create table Review_Events
%%sql 
DROP TABLE IF EXISTS Review_Events CASCADE;
CREATE TABLE Review_Events (
  eid int NOT NULL, 
  eoid int NOT NULL, 
  rid int, 
  PRIMARY KEY(rid), 
  FOREIGN KEY (eid, eoid) REFERENCES Event_Venue,
  FOREIGN KEY (rid) REFERENCES Reviews
)
```

```python
## create table Likes
%%sql 
DROP TABLE IF EXISTS Likes CASCADE;
CREATE TABLE Likes (
  uid int, 
  rid int, 
  PRIMARY KEY(rid, uid), 
  like_date date,
  FOREIGN KEY (uid) REFERENCES Users,
  FOREIGN KEY (rid) REFERENCES Reviews
)
```

```python
## create table Rate
%%sql 
DROP TABLE IF EXISTS Rate CASCADE;
CREATE TABLE Rate (
  uid int, 
  eid int, 
  eoid int, 
  star_count int, 
  PRIMARY KEY(uid, eid, eoid), 
  rate_date date,
  FOREIGN KEY (uid) REFERENCES Users,
  FOREIGN KEY (eid, eoid) REFERENCES Event_Venue,
  CHECK (
    star_count in (1, 2, 3, 4, 5)
  )
)
```

```python
## create table Share
%%sql 
DROP TABLE IF EXISTS Share CASCADE;
CREATE TABLE Share (
  uid int, 
  eid int, 
  eoid int,
  platform text, 
  PRIMARY KEY(uid, eid, eoid), 
  share_date date,
  FOREIGN KEY (uid) REFERENCES Users,
  FOREIGN KEY (eid, eoid) REFERENCES Event_Venue
)
```

```python
## create table Merchandise
%%sql 
DROP TABLE IF EXISTS Merchandise CASCADE;
CREATE TABLE Merchandise (
  mid int, 
  category text, 
  price float(2),
  PRIMARY KEY(mid)
)
```

```python
## create table User_Merchandise
%%sql 
DROP TABLE IF EXISTS User_Merchandise CASCADE;
CREATE TABLE User_Merchandise (
  uid int, 
  mid int, 
  PRIMARY KEY(uid, mid), 
  purchase_date date,
  FOREIGN KEY (uid) REFERENCES Users,
  FOREIGN KEY (mid) REFERENCES Merchandise
)
```

```python
## create table Merchandise_Events
%%sql 
DROP TABLE IF EXISTS Merchandise_Events CASCADE;
CREATE TABLE Merchandise_Events (
  eid int, 
  eoid int, 
  mid int, 
  PRIMARY KEY(mid, eid, eoid), 
  FOREIGN KEY (eid, eoid) REFERENCES Event_Venue,
  FOREIGN KEY (mid) REFERENCES Merchandise
)
```

## Some interesting queries

1. The name of the user who earn the most money by reselling his/her tickets.

```python
%%sql
SELECT u.user_name
FROM Users u 
WHERE uid IN
(SELECT uid
FROM
(SELECT r.uid, t.price - r.resell_price as price_diff 
FROM User_Buy_Tickets b 
INNER JOIN Tickets t 
ON b.tid = t.tid
INNER JOIN User_Resell_Tickets r 
ON b.tid = r.tid AND b.uid = r.uid
ORDER BY price_diff DESC
LIMIT 1) AS a1
)
 
```

\

2. The names of venues which hold more than one events

```python
%%sql
SELECT v2.name
FROM Venue v2
WHERE vid IN (
SELECT ev.vid
FROM Event_Venue ev
INNER JOIN Venue v 
ON ev.vid = v.vid
GROUP BY ev.vid
HAVING count(*) > 1)
```

\

3. The name of the artist whose events got the most resold tickets

```python
%%sql
SELECT name
FROM Artists a 
WHERE a.aid IN
(SELECT aid
FROM Events_Artists ea 
WHERE eid IN
(SELECT eid
FROM User_Resell_Tickets r
INNER JOIN Tickets t 
ON r.tid = t.tid
GROUP BY eid 
ORDER BY count(*) DESC
LIMIT 1)
)
```

4. The name of the event which has the most income for merchandise. 

```python
%%sql
SELECT events_name
FROM Event_Venue ev 
WHERE (eid, eoid) IN
(SELECT eid, eoid as sum
FROM Merchandise m
JOIN Merchandise_Events me 
ON m.mid = me.mid 
JOIN User_Merchandise um 
ON m.mid = um.mid 
GROUP BY eid, eoid
ORDER BY sum(price) DESC 
LIMIT 1)
```

5. The name of the event which has the lowest rate.

```python
%%sql
SELECT events_name
FROM Event_Venue ev 
WHERE (eid, eoid) IN
(SELECT eid, eoid
FROM Rate r
JOIN Event_Venue ev 
USING (eid, eoid) 
GROUP BY eid, eoid 
ORDER BY avg(star_count) 
LIMIT 1)
```

6. The name of the user who wrote more than 1 reviews.

```python
%%sql
SELECT user_name 
FROM Users 
WHERE uid IN
(SELECT uid
FROM User_Review ur
JOIN Users u
USING (uid) 
GROUP BY uid 
HAVING count(*) > 1)
```
